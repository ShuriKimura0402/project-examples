```python
from elevenlabs.client import ElevenLabs
from elevenlabs import save
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from IPython.display import Audio
```

```python
client = ElevenLabs(
    api_key="sk_9a2e7306ad1910e488bc88e477b536a305f894e740a2d90d"
)
```

```python
model_name = "sbintuitions/sarashina2.2-3b"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)
```

```python
prompt = "東京の魅力について話してください"
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
outputs = model.generate(
    **inputs,
    max_length=40,
    temperature=0.7,
    do_sample=True
)
```

```python
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
print(f"生成されたテキスト: {generated_text}")
```

```python
audio = client.text_to_speech.convert(
    voice_id="zDphKm8ApvCPgvBBZbPD",  #Manaka(私の声のクローンです)
    text=generated_text,
    model_id="eleven_multilingual_v2"
)
```
